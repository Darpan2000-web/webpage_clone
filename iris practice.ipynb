```python
df.plot(kind= 'scatter',x='SepalLengthCm',y='SepalWidthCm')
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue="Species" )\
.map(plt.scatter,"SepalLengthCm","SepalWidthCm").add_legend()
# This initializes a `FacetGrid` using your `df` DataFrame.
#The `hue="species"` argument tells seaborn to use the 'species' column
#to color-code the data points in the subsequent plots.
#This means that points belonging to different species will have
#different colors.
```

```python
df
```

```python
sns.set_style("whitegrid")
sns.pairplot(df,hue="Species")
plt.show()
```

```python
Iris_setosa = df.loc[df['Species'] =='iris-setosa']
Iris_versicolor = df.loc[df['Species']=='iris-versicolor']
Iris_virginica = df.loc[df['Species']=='iris-virginica']
```

```python
sns.FacetGrid(df,hue="Species")\
.map(sns.distplot,"PetalLengthCm")\
.add_legend()
plt.show()
```

```python
sns.FacetGrid(df,hue="Species")\
.map(sns.distplot,"PetalWidthCm")\
.add_legend()
plt.show()
```

```python
sns.FacetGrid(df,hue="Species",height=6)\
.map(sns.distplot,"PetalWidthCm")
```

```python
# Box Plot
 #Box-plot can be visualized as a PDF on the side-ways.
sns.boxplot(x='Species',y='PetalLengthCm', data=df)
plt.show()
```
